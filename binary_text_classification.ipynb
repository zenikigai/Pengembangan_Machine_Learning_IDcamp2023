# Binary Text Classification

Imagine if you own a restaurant and you have hundreds of thousands of customer reviews. You want the distinction between positive and negative reviews to be automated so you can easily filter customer reviews. In this submodule we will exercise for this problem.

The dataset consists of 2 classes namely 0 and 1 which indicate whether the review is positive or negative.

Download the dataset here: https://www.kaggle.com/datasets/marklvl/sentiment-labelled-sentences-data-set

```python
# We create a dataframe from our dataset to make it easier to process.

import pandas as pd
df = pd.read_csv("yelp_labelled.txt", names=["sentence", "label"], sep="\t")
```

```python
# show the last 5 samples fromour dataset
df.tail()
```

```python
# Divide the dataset into train and test sets.
from sklearn.model_selection import train_test_split
sentences = df["sentence"].values
y = df["label"].values
train_sentences, test_sentences, y_train, y_test = train_test_split(sentences, y, test_size=0.2)
```

In order for the text to be understood by the model, we must tokenize it. Use the tokenizer function on both the train data and the test data. Don't forget to use the pad_sequences function so that each sequence is the same length.

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words = 250, oov_token="x")
tokenizer.fit_on_texts(train_sentences)
tokenizer.fit_on_texts(test_sentences)

train_sequence = tokenizer.texts_to_sequences(train_sentences)
test_sequence = tokenizer.texts_to_sequences(test_sentences)

padded_train = pad_sequences(train_sequence, maxlen=20)
padded_test = pad_sequences(test_sequence, maxlen=20)
```

The architecture that will be used is layer embedding, with the first argument corresponding to the number of vocabulary / words we use in the tokenizer. The next argument is the embedding dimension, and the input_length which is the length of the sequence. Well we don't use a Flatten layer but we replace it with GlobalAveragePooling1D. This function works better in NLP cases than Flatten.

```python
from keras.models import Sequential
from keras.layers import Embedding, GlobalAveragePooling1D, Dense

model = Sequential([
    Embedding(250, 16, input_length=20),
    GlobalAveragePooling1D(),
    Dense(24, activation="relu"),
    Dense(1, activation="sigmoid")
])
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# train the model
num_epochs = 100
history = model.fit(padded_train, y_train, epochs=num_epochs,
                    validation_data=(padded_test, y_test),
                    verbose=1)
```
