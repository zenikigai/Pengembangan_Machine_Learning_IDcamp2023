Implementing callback excercise. Callback is a way to stop the model training when the model reach certain accuracy point. We can do that with callback function. We take the previous code https://github.com/zenikigai/Pengembangan_Machine_Learning_IDcamp2023/blob/main/Jst_multi_class.ipynb and try to implement callback here.

```python
# import library we added tensorflow
import tensorflow as tf
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```python
# change the dataset to be pandas dataframe
df = pd.read_csv("Iris.csv")
df
```

```python
# delete (drop()) the Id column, we don't need to use it
df = df.drop(columns = "Id")
```

```python
# one hot encoding, because our label is categorical
category = pd.get_dummies(df.Species)
category
```

```python
# We need to merge the result column of one hot encoding
# remove the species column because that column will not be used.
new_df = pd.concat( [df, category], axis=1 )
new_df = new_df.drop(columns="Species")
new_df
```

```python
# convert those dataframe into numpyt array with "values" function from the dataframe
dataset = new_df.values
dataset
```

```python

# separate the attribute and the label
# choose the first 4 columns as attribute
X = dataset[:, 0:4]

# choose the last 3 columns as the label
y = dataset[:, 4:7]
```

```python
# Normalization
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
X_scale
```

```python
# divide the data to be training data and test data, we give 30% for the test data
X_train, X_test, Y_train, Y_test = train_test_split(X_scale, y, test_size=0.3)
```

```python
# determine the model architecture, i use 3 layers, activation func "softmax" because it's multiclass
model = Sequential([
    Dense(64, activation="relu", input_shape=(4,)),
    Dense(64, activation="relu"),
    Dense(3, activation="softmax"),
])
```

```python
# determine optimizer and loss function, for this multiclass lesson, choose "categorical_crossentropy"
model.compile(optimizer="Adam",
              loss="categorical_crossentropy",
              metrics=["accuracy"])
```

```python
# making callback class
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get("accuracy") > 0.9):
      print("\nAccuracy have reach >90%!")
      self.model.stop_training = True
callbacks = myCallback()
```

 function on_epoch_end() is the function that will tell the model to stop the training when certain target accuracy is fulfill.

 next is the if(logs.get(‘accuracy’)>0.9), it means that "if the accuracy bigger that 0.9, execute the next command" which is

 print("\nAccuracy have reach >90%!"), will be execute when callback active.

 self.model.stop_training = True, is the code that will tell the model to stop the training.

```python
# use the callback class
model.fit(X_train, Y_train, epochs=50, callbacks=[callbacks])
```
